```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have already loaded the image as 'nemo'
# nemo = cv2.imread('path_to_your_image.jpg')

# Convert the color space from BGR to RGB
nemo = cv2.cvtColor(nemo, cv2.COLOR_BGR2RGB)

# Define the black threshold (adjust these values based on your desired range)
lower_black = np.array([0, 0, 0])      # Lower bound for black (RGB values: [R, G, B])
upper_black = np.array([50, 50, 50])   # Upper bound for black (RGB values: [R, G, B])

# Function to turn black pixels within the specified range to white
def change_black_range_to_white(image, lower_threshold, upper_threshold):
    # Create a boolean mask to identify the black pixels within the specified range
    mask = np.all((image >= lower_threshold) & (image <= upper_threshold), axis=-1)

    # Set the pixels identified by the mask to white
    image[mask] = [255, 255, 255]

    return image

# Apply the function to turn the specific range of black pixels to white
nemo_changed_black_range = change_black_range_to_white(nemo, lower_black, upper_black)

# Display the modified image
plt.imshow(nemo_changed_black_range)
plt.show()
```

```python
import cv2
import pandas as pd
import random
import matplotlib.pyplot as plt

path = "/home/ec2-user/SageMaker/dataset/uncropped_300/"
annotations_file = "/home/ec2-user/SageMaker/dataset/uncropped/annotations.csv"

# Read the annotations CSV file
annotations_df = pd.read_csv(annotations_file)

# Filter images where the label is 1
nemos_friends = []
for idx, row in annotations_df.iterrows():
    if row['Label'] == 1:
        friend = cv2.cvtColor(
            cv2.imread(path + row['ImageName']), cv2.COLOR_BGR2RGB
        )
        nemos_friends.append(friend)

# Randomly pick 6 images
random_friends = random.sample(nemos_friends, 10)

def segment_fish(image):
    """Attempts to segment the clown fish out of the provided image."""
    hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    light_orange = (0.0000, 0.0000, 27.4510)
    dark_orange = (0.0000, 0.0000, 125.8235)
    mask = cv2.inRange(hsv_image, light_orange, dark_orange)
    final_mask = mask
    result = cv2.bitwise_and(image, image, mask=final_mask)
    # result = cv2.GaussianBlur(result, (7, 7), 0)
    return result

results = [segment_fish(friend) for friend in random_friends]

# Display the images
for i in range(10):
    plt.subplot(1, 2, 1)
    plt.imshow(random_friends[i])
    plt.title("Original Image")
    plt.subplot(1, 2, 2)
    plt.imshow(results[i])
    plt.title("Segmented Image")
    plt.show()
```

```python
import cv2
import numpy as np

def segment_image_using_gray(image):
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, binary_image = cv2.threshold(gray_image, 128, 255, cv2.THRESH_BINARY)
    return binary_image

def segment_image_using_hsv(image):
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    value_channel = hsv_image[:, :, 2]
    _, binary_image = cv2.threshold(value_channel, 128, 255, cv2.THRESH_BINARY)
    return binary_image

def segment_image_using_lab(image):
    lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    l_channel = lab_image[:, :, 0]
    _, binary_image = cv2.threshold(l_channel, 128, 255, cv2.THRESH_BINARY)
    return binary_image

if __name__ == "__main__":
    # Using a smaller test image for demonstration
    image_path = "dataset/uncropped/soloAUV 20220919-163749-005.png"
    original_image = cv2.imread(image_path)

    if original_image is None:
        print("Error: Image not found or cannot be read.")
        exit()

    try:
        segmented_gray = segment_image_using_gray(original_image)
        segmented_hsv = segment_image_using_hsv(original_image)
        segmented_lab = segment_image_using_lab(original_image)

        # Display the segmented images
        cv2.imshow("Original Image", original_image)
        cv2.imshow("Segmented Gray", segmented_gray)
        cv2.imshow("Segmented HSV", segmented_hsv)
        cv2.imshow("Segmented LAB", segmented_lab)

        cv2.waitKey(0)
    except Exception as e:
        print("Error occurred:", str(e))
    finally:
        cv2.destroyAllWindows()
```

```python
import os
import cv2
import pandas as pd
import numpy as np
import re

def create_mask(image_shape, polygon_coords_list):
    mask = np.zeros(image_shape[:2], dtype=np.uint8)
    for polygon_coords in polygon_coords_list:
        cv2.fillPoly(mask, [polygon_coords], 255)  # Set pixel value to 255 (white) for object pixels
    return mask

def extract_coordinates(coordinate_str):
    # Check if coordinate_str is NaN or empty
    if pd.isna(coordinate_str) or not coordinate_str:
        return np.array([])

    # Extract coordinates from the string using regular expression
    coordinate_list = re.findall(r'\d+', coordinate_str)
    # Convert to a list of integers
    coordinates = list(map(int, coordinate_list))
    # Reshape to pairs of (x, y) coordinates
    return np.array(coordinates).reshape(-1, 2)

def main(input_folder, csv_file, output_folder):
    # Create the output folders if they don't exist
    os.makedirs(output_folder, exist_ok=True)
    os.makedirs(os.path.join(output_folder, "images"), exist_ok=True)
    os.makedirs(os.path.join(output_folder, "masks"), exist_ok=True)

    # Read the CSV file
    df = pd.read_csv(csv_file)

    image_objects = {}  # Dictionary to store polygon coordinates for each image

    for index, row in df.iterrows():
        filename = os.path.join(input_folder, row['ImageName'])

        # Check if the 'Repeated Species' column value is "marine-snow"
        if row['Repeated Species'] == "marine-snow":
            print(f"Skipping {filename} as the 'Repeated Species' is 'marine-snow'.")
            continue

        image = cv2.imread(filename)
        polygon_coords = []
        # If the 'poly_coordinates' column is empty or has no data, create a polygon from TL and BR coordinates
        if pd.isna(row['poly_coordinates']) or not row['poly_coordinates']:
            TL_x, TL_y = row['TL_x'], row['TL_y']
            BR_x, BR_y = row['BR_x'], row['BR_y']
            cords = [(TL_x, TL_y), (TL_x, BR_y), (BR_x, BR_y), (BR_x, TL_y)]
            polygon_coords = np.array(cords, np.int32)
        else:
            # Extract polygon coordinates from the string
            polygon_coords_str = row['poly_coordinates']
            polygon_coords = extract_coordinates(polygon_coords_str)

        # Get the base filename (without extension) from the original image
        base_filename = os.path.splitext(os.path.basename(filename))[0]

        # Accumulate the polygon coordinates for each image
        if base_filename not in image_objects:
            image_objects[base_filename] = []
        image_objects[base_filename].append(polygon_coords)

        # Save the resized image and mask only if all entries have valid polygon coordinates
        if index == len(df) - 1 or row['ImageName'] != df.loc[index + 1, 'ImageName']:
            # Check if at least one set of polygon coordinates is present for this image
            if len(image_objects[base_filename]) > 0:
                # Load the original image (without any edits)
                original_image = cv2.imread(filename)

                # Resize the image to 732x732
                resized_image = cv2.resize(original_image, (512, 512))

                # Create the mask based on the accumulated polygon coordinates
                mask = create_mask(original_image.shape, image_objects[base_filename])

                # Resize the mask to 732x732
                resized_mask = cv2.resize(mask, (512, 512))

                # Save the resized image to the output folder
                output_image_filename = os.path.join(output_folder, "images", f"{base_filename}.png")
                cv2.imwrite(output_image_filename, resized_image)

                # Save the resized mask to the output folder
                output_mask_filename = os.path.join(output_folder, "masks", f"{base_filename}_mask.png")
                cv2.imwrite(output_mask_filename, resized_mask)

            # Clear the polygon coordinates for the current image
            image_objects.pop(base_filename, None)

if __name__ == "__main__":
    input_folder = "/home/ec2-user/SageMaker/dataset/All_Images/2022-09-19/"
    csv_file = "/home/ec2-user/SageMaker/OpenCV/csv/19.csv"
    output_folder = "/home/ec2-user/SageMaker/OpenCV/try/"  # Separate folder for resized images and masks
    main(input_folder, csv_file, output_folder)
```

# Visualization

```python
import os
import cv2
import numpy as np

def overlay_mask(image, mask):
    mask_rgb = cv2.merge([mask, mask, mask])
    alpha = 0.3  # Reduce the intensity of the overlay
    overlay = cv2.addWeighted(image, 1 - alpha, mask_rgb, alpha, 0)
    return overlay

def visualize_image_mask(image_folder, mask_folder, image_filename, resize_shape=(256, 256)):
    # Load the specified image and its corresponding mask
    image_filepath = os.path.join(image_folder, image_filename)
    mask_filename = f"{os.path.splitext(image_filename)[0]}.png"
    mask_filepath = os.path.join(mask_folder, mask_filename)

    if not os.path.exists(image_filepath) or not os.path.exists(mask_filepath):
        print(f"Error: Image or mask file not found for {image_filename}")
        return

    image = cv2.imread(image_filepath)
    mask = cv2.imread(mask_filepath, cv2.IMREAD_GRAYSCALE)

    # Resize the image and mask
    image = cv2.resize(image, resize_shape)
    mask = cv2.resize(mask, resize_shape)

    # Overlay the mask on the image
    overlay = overlay_mask(image, mask)

    # Display the images using matplotlib (optional)
    import matplotlib.pyplot as plt

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 3, 1)
    plt.title("Resized Image")
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title("Resized Mask")
    plt.imshow(mask, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.title("Image with Overlay")
    plt.imshow(cv2.cvtColor(overlay, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    image_folder = "/home/ec2-user/SageMaker/OpenCV/resized_data/images/"
    mask_folder = "/home/ec2-user/SageMaker/OpenCV/resized_data/masks/"
    image_filename = "soloAUV 20220908-161939-004.png"
    visualize_image_mask(image_folder, mask_folder, image_filename)
```

```python
import cv2
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image
import random

# Load the image and mask
image = cv2.imread("/home/ec2-user/SageMaker/OpenCV/512_final/images/soloAUV 20220908-183646-070.png")
mask = cv2.imread("/home/ec2-user/SageMaker/OpenCV/512_final/masks/soloAUV 20220908-183646-070_mask.png", 0) # Load the mask as grayscale (single channel)
image_pil = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
mask_pil = Image.fromarray(mask)

# Define the transformations
random_transforms = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(degrees=20)
])

# Generate a random seed for each image-mask pair
seed = random.randint(0, 1000)

# Apply the random transformations to the image and the mask using the generated seed
torch.manual_seed(seed)
transformed_image = random_transforms(image_pil)
torch.manual_seed(seed)  # Reset the seed for applying the same transformations to the mask
transformed_mask = random_transforms(mask_pil)

# Convert the transformed mask to a tensor
transformed_mask_tensor = transforms.ToTensor()(transformed_mask)

# Visualize the transformed image and mask
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Transformed Image")
plt.imshow(transformed_image)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Transformed Mask")
plt.imshow(transformed_mask_tensor[0], cmap='gray')
plt.axis('off')

plt.show()
```

```python
import os
import cv2
import matplotlib.pyplot as plt
from torchvision import transforms
from PIL import Image
import random

# Function to apply random transformations and save augmented images and masks
def augment_and_save(image_path, mask_path, output_dir):
    # Load the image and mask
    image = cv2.imread(image_path)
    mask = cv2.imread(mask_path, 0) # Load the mask as grayscale (single channel)
    image_pil = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    mask_pil = Image.fromarray(mask)

    # Define the transformations
    random_transforms = transforms.Compose([
        transforms.RandomHorizontalFlip(),
        transforms.RandomVerticalFlip(),
        transforms.RandomRotation(degrees=20)
    ])

    # Generate a random seed for each image-mask pair
    seed = random.randint(0, 1000)

    # Apply the random transformations to the image and the mask using the generated seed
    torch.manual_seed(seed)
    transformed_image = random_transforms(image_pil)
    torch.manual_seed(seed)  # Reset the seed for applying the same transformations to the mask
    transformed_mask = random_transforms(mask_pil)

    # Convert the transformed mask to a tensor
    transformed_mask_tensor = transforms.ToTensor()(transformed_mask)

    # Save the augmented image and mask
    image_name = os.path.basename(image_path)
    mask_name = os.path.basename(mask_path)
    augmented_image_path = os.path.join(output_dir, "images_agumented", image_name.replace(".png", "_agumented.png"))
    augmented_mask_path = os.path.join(output_dir, "masks_agumented", mask_name.replace(".png", "_agumented_mask.png"))

    os.makedirs(os.path.dirname(augmented_image_path), exist_ok=True)
    os.makedirs(os.path.dirname(augmented_mask_path), exist_ok=True)

    transformed_image.save(augmented_image_path)
    transformed_mask_tensor_np = transformed_mask_tensor.numpy()
    cv2.imwrite(augmented_mask_path, transformed_mask_tensor_np[0] * 255)

# Define the directories for images and masks
images_directory = "/home/ec2-user/SageMaker/OpenCV/512_final/images/"
masks_directory = "/home/ec2-user/SageMaker/OpenCV/512_final/masks/"
output_directory = "/home/ec2-user/SageMaker/OpenCV/512_final/"

# Get the list of image files in the images directory
image_files = os.listdir(images_directory)

# Iterate over the image files and perform augmentation
for image_file in image_files:
    if image_file.endswith(".png"):
        image_path = os.path.join(images_directory, image_file)
        mask_file = image_file.replace(".png", "_mask.png")
        mask_path = os.path.join(masks_directory, mask_file)

        # Perform augmentation and save the results
        augment_and_save(image_path, mask_path, output_directory)

print("Augmentation and saving completed.")
```
